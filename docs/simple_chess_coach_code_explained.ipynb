## Part 1 — Backend API (`api_main.py`)

The API is the part of the system that receives input and returns recommendations.

In this prototype, the API is built with **FastAPI**.

FastAPI is useful because it:
- is written in Python
- is easy to test
- is simple to connect to a front-end
- makes the project look more like a real software system

````python
# Example backend structure from the prototype

from fastapi import FastAPI
import chess

app = FastAPI()

@app.get("/")
def root():
    return {"message": "Chess Coach API running"}

@app.post("/recommend-move")
def recommend_move(fen: str):
    board = chess.Board(fen)
    moves = list(board.legal_moves)[:5]
    return {"moves": [str(m) for m in moves]}
````

### What this backend code is doing

1. **Creates the FastAPI app**
   - `app = FastAPI()`
   - this initializes the API service

2. **Defines a basic route**
   - `@app.get("/")`
   - this is a simple test route to verify the server is running

3. **Defines the recommendation route**
   - `@app.post("/recommend-move")`
   - this accepts a chess position and returns move suggestions

4. **Uses `python-chess`**
   - `chess.Board(fen)` loads the board
   - `board.legal_moves` gets legal moves from that position

5. **Returns results as JSON**
   - that makes it easy for the UI to display recommendations

## Why this matters

This satisfies the **API requirement** of the capstone.

Even in simplified form, the backend shows that the project is not just a static mockup. It is an actual system that accepts input and produces output through an application programming interface.

## Part 2 — Front-End UI (`streamlit_app.py`)

The UI is what teammates and users actually see.

In this prototype, the UI is built with **Streamlit**.

Streamlit is a strong choice for a prototype because:
- it is fast to build
- it uses Python
- it is easy to demo
- it creates a clean interface without requiring a full React front-end

````python
# Example front-end structure from the prototype

import streamlit as st
import requests

st.title("Adaptive Chess Coach")

fen = st.text_input("Enter FEN")

if st.button("Analyze"):
    r = requests.post("http://127.0.0.1:8000/recommend-move", params={"fen": fen})
    st.write(r.json())
````

### What this UI code is doing

1. **Imports Streamlit**
   - Streamlit builds the page and input fields

2. **Imports `requests`**
   - `requests` is how the UI talks to the backend API

3. **Shows a title**
   - helps the page look like an application

4. **Accepts user input**
   - in older versions this was FEN
   - in the simpler version this can be replaced by easier options like pasted moves or sample positions

5. **Sends the data to the API**
   - the UI does not do all the thinking itself
   - it passes the data to the backend

6. **Displays the API response**
   - the user sees the recommended moves or explanation output
